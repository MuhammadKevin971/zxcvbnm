```python
import pandas as pd

file_path = r"C:\Users\ericl\Downloads\opendata.jabarprov.go.id_dataset_od_16985_jumlah_produksi_sampah_berdasarkan_kabupatenkota_v3_xls\data.jabarprov.go.id\disperkim-od_16985_jumlah_produksi_sampah_berdasarkan_kabupatenkota_v3_data.xlsx"
df = pd.read_excel(file_path)

df['jumlah_produksi_sampah'] = pd.to_numeric(df['jumlah_produksi_sampah'], errors='coerce')

data_sampah_df = pd.DataFrame({
    'Lokasi': df['nama_kabupaten_kota'],
    'Sampah (Ton/Hari)': df['jumlah_produksi_sampah'],
    'Tahun': df['tahun']
})

print(data_sampah_df)

tahun_tertentu = 2019
total_sampah = 0

for index, row in data_sampah_df.iterrows():
    if row['Tahun'] == tahun_tertentu:
        total_sampah += row['Sampah (Ton/Hari)']

print(f"\nTotal Produksi Sampah di Jawa Barat untuk tahun {tahun_tertentu}: {total_sampah:.2f} Ton/Hari")

total_per_tahun = {}
for index, row in data_sampah_df.iterrows():
    if row['Tahun'] not in total_per_tahun:
        total_per_tahun[row['Tahun']] = 0
    total_per_tahun[row['Tahun']] += row['Sampah (Ton/Hari)']

print("\nTotal Produksi Sampah per Tahun:")
for tahun, total in total_per_tahun.items():
    print(f"Tahun {tahun}: {total:.2f} Ton/Hari")

total_per_kabupaten_per_tahun = {}
for index, row in data_sampah_df.iterrows():
    key = (row['Lokasi'], row['Tahun'])
    if key not in total_per_kabupaten_per_tahun:
        total_per_kabupaten_per_tahun[key] = 0
    total_per_kabupaten_per_tahun[key] += row['Sampah (Ton/Hari)']

print("\nTotal Produksi Sampah per Kota/Kabupaten per Tahun:")
for (lokasi, tahun), total in total_per_kabupaten_per_tahun.items():
    print(f"{lokasi} - Tahun {tahun}: {total:.2f} Ton/Hari")

data_sampah_df.to_csv('total_produksi_sampah.csv', index=False)

data_sampah_df.to_excel('total_produksi_sampah.xlsx', index=False)

total_per_tahun_df = pd.DataFrame(list(total_per_tahun.items()), columns=['Tahun', 'Total Sampah (Ton/Hari)'])
total_per_tahun_df.to_csv('total_per_tahun.csv', index=False)

total_per_kabupaten_per_tahun_df = pd.DataFrame(list(total_per_kabupaten_per_tahun.items()), columns=['Lokasi-Tahun', 'Total Sampah (Ton/Hari)'])
total_per_kabupaten_per_tahun_df.to_csv('total_per_kabupaten_per_tahun.csv', index=False)
```
